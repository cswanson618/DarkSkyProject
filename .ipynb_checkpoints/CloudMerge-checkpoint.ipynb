```python
import pandas as pd
import numpy as np
import requests
import json
import os
import pprint
# from statesjson import us_states_json
# from states import us_states
from citipy import citipy
```

```python
weather_df = pd.read_csv("his_weather_final.csv")
PC = pd.read_csv("Park Coordinates.csv")
pcrounded = PC.round({"Latitude": 4, "Longitude": 3})
```

```python
def cartesian_product_basic(left, right):
    return (
       weather_df.assign(key=1).merge(pcrounded.assign(key=1), on='key').drop('key', 1))
weather_merged_df = cartesian_product_basic(weather_df, pcrounded)
#Credit: Umair Khakoo
```

```python
weather_df
```

```python
weather_df_unclean = weather_merged_df[
    ((abs(weather_merged_df["Latitude_x"] - weather_merged_df["Latitude_y"])) <= 0.44) &\
    ((abs(weather_merged_df["Longitude_x"] - weather_merged_df["Longitude_y"])) <= 0.44)
]
```

```python
weather_df_unclean
```

```python
weather_no2020_unclean = weather_df_unclean[weather_df_unclean["Date time"] != "01/01/2020"]
weather_wanted_unclean = weather_no2020_unclean.drop(columns = {"Address", "Minimum Temperature", "Maximum Temperature", "Temperature", "Dew Point", "Relative Humidity", "Heat Index", "Wind Speed", "Wind Gust", "Sea Level Pressure", "Weather Type", "Name", "Info", "Wind Direction", "Wind Chill", "Precipitation", "Precipitation Cover", "Resolved Address", "Snow Depth", "Visibility"})
```

```python
weather_dates = pd.to_datetime(weather_wanted_unclean["Date time"])
weather_date_df = pd.DataFrame(weather_dates)
weather_date_drop = weather_wanted_unclean.drop(columns= "Date time")
weather_df_clean = weather_date_drop.merge(weather_date_df, left_index = True, right_index = True)
```

```python
weather_df_date_group = weather_df_clean.set_index(["Date time"])
```
