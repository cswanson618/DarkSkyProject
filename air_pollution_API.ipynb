```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from pprint import pprint
pd.options.display.max_columns = 9999
```

```python
# Import park coordinates csv file as pandas dataframe
df = pd.read_csv("park_coordinates.csv")
df["Min Lat"] = df["Latitude"] - 0.44
df["Max Lat"] = df["Latitude"] + 0.44
df["Min Lon"] = df["Longitude"] - 0.44
df["Max Lon"] = df["Longitude"] + 0.44
df = round(df, 2)
df
```

```python
# Air Quality Data Query

url = "https://aqs.epa.gov/data/api/annualData/byBox?email=test@aqs.api&key=test&param=44201&bdate=20150608&edate=20151020"
aq_data = []

for i in range(0, len(df)-1):
    result = requests.get(url + 
                     "&minlat=" + str(df.iat[i, 3]) + 
                     "&maxlat=" + str(df.iat[i, 4]) + 
                     "&minlon=" + str(df.iat[i, 5]) + 
                     "&maxlon=" + str(df.iat[i, 6])).json()
    if result["Data"]==[]:
        print(str(df.iloc[i, 2]) + ": No site nearby")
    else:
        aq_data.append(result)
```

```python
# Getting values for only "Data"

aq_data_clean = []
for d in aq_data:
    aq_data_clean += d["Data"]
```

```python
# Converting to Pandas dataframe

df_aq = pd.DataFrame(aq_data_clean)
df_aq
```
